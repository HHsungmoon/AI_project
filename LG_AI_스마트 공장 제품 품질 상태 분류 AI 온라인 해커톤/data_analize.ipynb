```python
"""
csv 파일의 null 부분을 확인해 봤더니 "X_1 ~ X_92 423" 처럼 null 인자 갯수가 같은 target 값들이 눈에 띄었다.
예를들어 train_22, train_23은 null, null이 아닌 target 종류가 같았다. 
아래 코드는 target마다 null 타입의 갯수를 모아 같은 갯수끼리 그룹화 해본 것이다.
"""

list_null_check = df.isnull().sum()
print(type(list_null_check))

print(list_null_check.index[0])

"""
for i in range(0, len(list_null_check)):
    if list_null_check[i] != 0:  
        print(list_null_check.index[i], " : " , list_null_check[i])      
"""    
tmp = 0
str_tmp = ""
start_idx = list_null_check.index[6]
dict = {}
tmp = list_null_check[0]
for i in range(6, len(list_null_check)):
    if tmp != list_null_check[i]:
        tmp = list_null_check[i]
        end_idx = list_null_check.index[i]
        s = start_idx + " ~ " + end_idx
        start_idx = end_idx
        dict[s] = tmp
        s = ""
        
for key, value in dict.items():
    print(key, value)
   
```

```python
"""
y class 종류는 0,1,2 3가지 인데 Y_Quality의 값의 범위에 따라 나뉘어 지는 것이 확인된다. 
다만 Y_Quality의 값 자체의 차이가 적은 상태이다.
"""

Y_class_0 = []
Y_class_1 = []
Y_class_2 = []

for i in range(0, 597):
    if df["Y_Class"][i] == 0:
        Y_class_0.append(df["Y_Quality"][i])
    if df["Y_Class"][i] == 1:
        Y_class_1.append(df["Y_Quality"][i])
    if df["Y_Class"][i] == 2:
        Y_class_2.append(df["Y_Quality"][i])
        
print("class_0 : {:.8f} ~ {:.8f}".format(min(Y_class_0), max(Y_class_0)))
print("class_1 : {:.8f} ~ {:.8f}".format(min(Y_class_1), max(Y_class_1)))
print("class_2 : {:.8f} ~ {:.8f}".format(min(Y_class_2), max(Y_class_2)))
```

```python
# df의 null을 -1로 채우는 방법
df.fillna(-1, inplace=True)

label_X = [[]]

for idx in range(1, 2875):
    label_name = "X_" + str(idx)
    input_list = []
    for i in range(0,597):
        if df[label_name][i] != -1:
            input_list.append(df[label_name][i])
    label_X.append(input_list)
        
```
